```python
"""
DISEASE MODULE COHESION RANKING

This script ranks genes according to their disease module cohesion.

Requirements:
rankedonlyGWAS: the ranked and annotated matrix of genes of interest
degreefile: a file with degrees of all nodes in the chosen network. An example file of degrees in the interactome is provided: interactome_degrees.txt
outputfile: the results file 

This script should be run in a jupyter notebook
"""
#
#
# trying things in the pandas way 
#
import pandas as pd
ranked_matrix=pd.read_csv(rankedonlyGWAS, sep='\t')
#print(ranked_matrix.head())
novel_genes=ranked_matrix[ranked_matrix.knowngene=="novelgene"]
#print(novel_genes.head())

#degrees
degree_matrix=pd.read_csv(degreefile, sep='\t')
#print(degree_matrix.head())
#dict with degrees
d_degree = dict(zip(degree_matrix.node, degree_matrix.degree))

d_final_rank={}
d_degrees_of_candiadtes={}
d_hubtype={}
for index, rows in ranked_matrix.iterrows(): 
    mysym= rows.symbol
    ID= rows.entrez_ID
    d_degrees_of_candiadtes[ID]=d_degree[ID]
    if d_degree[ID]>=100:
        hubtype='hub'
    else:
        hubtype='norm'
    d_hubtype[ID]=hubtype
    my_list =[rows.visit_probab_r09_rank, rows.visit_probab_r08_rank, rows.visit_probab_r07_rank, rows.visit_probab_r06_rank
             , rows.visit_probab_r05_rank, rows.visit_probab_r04_rank, rows.visit_probab_r03_rank, rows.visit_probab_r02_rank, rows.visit_probab_r01_rank] 
    flag = 0
    i = 1
    while i < len(my_list): 
        if(my_list[i] < my_list[i - 1]): 
            flag = 1
        i += 1
    # printing result 
    if (not flag) :
        if rows.visit_probab_r09_rank< rows.visit_probab_r01_rank:
            if rows.knowngene=='novelgene':
                d_final_rank[ID]='candidate'
            if rows.visit_probab_r09_rank<44:
                if rows.knowngene=='novelgene':
                    #if d_degree[ID]<100:
                    d_final_rank[ID]='TOP_candidate'
                    print('top candidate: %s' %mysym)
    else : 
        continue
        d_final_rank[ID]='no'

new_d = pd.Series(d_final_rank)
new_d_2 = pd.Series(d_degrees_of_candiadtes)

#adding info
ranked_matrix['type']= ranked_matrix['entrez_ID'].map(d_final_rank)
ranked_matrix['degree']= ranked_matrix['entrez_ID'].map(d_degrees_of_candiadtes)
ranked_matrix['hubtype']= ranked_matrix['entrez_ID'].map(d_hubtype)
ranked_matrix.to_excel(outputfile)
print(ranked_matrix.head())

```
